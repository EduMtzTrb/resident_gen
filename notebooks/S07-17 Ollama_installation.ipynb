## 1. Download Ollama:

https://ollama.com/download?utm_source=chatgpt.com

## 2. Install Ollama locally

Go to the Ollama download page and install it for your OS: macOS, Windows, or Linux. For Linux, the official command is:
```Bash
curl -fsSL https://ollama.com/install.sh | sh
```
For macOS/Windows, you can use the installer from Ollama website.

installation_01.png

## 3. Check that Ollama works

Open Terminal / PowerShell:
```Bash
ollama --version
```
Expected result: it prints the Ollama version.

## 4. Download Qwen3 8B locally
```Bash
ollama pull qwen3:8b
```
This downloads the model to the personal computer.

## 5. Test the model in terminal

```Bash
ollama run qwen3:8b
```
Then type:

```Bash
Hello, explain GraphRAG in one sentence.
```
Expected result: the model answers in the terminal.

To exit:

```Bash
/bye
```

## 6. Optional: use it in Jupyter Notebook

Install the Python client:

```Bash
pip install notebook ollama
```

In a notebook cell:

```python
import ollama

response = ollama.chat(
    model="qwen3:8b",
    messages=[
        {"role": "user", "content": "Hello, explain GraphRAG in one sentence."}
    ]
)

print(response["message"]["content"])
```



## 7. You need to start the Ollama service after restart:

Open Terminal / PowerShell
Run:
```Bash
ollama serve
```
