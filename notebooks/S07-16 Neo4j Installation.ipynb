## Step 7 — Connect to the database from Python

Paste your **Connection URL**, **username**, and **password** from the downloaded credentials file into the cell below.

```python
from neo4j import GraphDatabase

# Paste your values from the downloaded credentials file
URI      = "neo4j+s://<your-instance-id>.databases.neo4j.io"
USERNAME = "neo4j"
PASSWORD = "<your-password>"

driver = GraphDatabase.driver(URI, auth=(USERNAME, PASSWORD))
driver.verify_connectivity()
print("Connected successfully!")

---
## Step 8 — Run a test query

Verify the connection works by running a simple Cypher query that returns the Neo4j version.

```python

with driver.session() as session:
    result = session.run("RETURN 'Hello, Neo4j!' AS message, neo4j.version() AS version")
    for record in result:
        print(record["message"])
        print("Neo4j version:", record["version"])
